```python

from scipy import stats

z_scores = np.abs(stats.zscore(data))

# Define a threshold for outlier detection (e.g., z-score > 3)
threshold = 3

# Find outlier indices for the entire DataFrame
outlier_indices = np.where(z_scores > threshold)

# Flatten and get unique outlier indices
outlier_indices = np.unique(outlier_indices[0])

# Remove outliers from the DataFrame
data = data.drop(outlier_indices)

data
```

# Normalization basically Standarization

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
```

```python
X_train_std=std.fit_transform(X_train)
X_test_std=std.transform(X_test)
```

```python
import pickle
with open('std.pkl', 'wb') as f:
    pickle.dump(std, f)
```

```python
X_train
```

```python
X_train_std
```

```python
Y_test
```

# Model training

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train_std,Y_train)
```

```python
Y_pred=lr.predict(X_test_std)
```

```python
Y_pred
```

```python
Y_test
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
mean_absolute_error(Y_test,Y_pred)
```

```python
X_test
```

```python
X_test.loc[14444]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Assuming y_true contains the actual values and y_pred contains the predicted values
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
rmse = mean_squared_error(Y_test, Y_pred, squared=False)  # Compute RMSE
r_squared = r2_score(Y_test, Y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared Score:", r_squared)
```

## using other models and checking the r2_score

```python
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import r2_score

# Lasso Regression
lasso = Lasso(alpha=0.1)  # Adjust the alpha parameter as needed
lasso.fit(X_train_std, Y_train)
Y_pred_lasso = lasso.predict(X_test_std)
r2_score_lasso = r2_score(Y_test, Y_pred_lasso)

# Ridge Regression
ridge = Ridge(alpha=0.1)  # Adjust the alpha parameter as needed
ridge.fit(X_train_std, Y_train)
Y_pred_ridge = ridge.predict(X_test_std)
r2_score_ridge = r2_score(Y_test, Y_pred_ridge)

print("R-squared score for Lasso Regression:", r2_score_lasso)
print("R-squared score for Ridge Regression:", r2_score_ridge)
```
